```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement du jeu de données de test
test_data = pd.read_csv('atp_tennis_feature.csv')
```

```python
# Séparation des fonctionnalités et de la variable cible
X_test = test_data.drop('Winner', axis=1)
y_test = test_data['Winner']
```

```python
#Chargement du modèle
from joblib import load
model = load('atp_tennis_model.joblib')

# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test)
```

```python
# Evaluation du modèle
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
```

```python
# Affichage des résultats
print(f"Accuracy: {accuracy:2f}")
print("Classification Report:")
print(report)
```

```python
# Affichage de la matrice de confusion
plt.figure(figsize=(5, 4))
sns.heatmap(confusion, annot=True, fmt="d", cmap='Blues', cbar=False)
plt.xlabel('Predictions')
plt.ylabel('Vraies valeurs')
plt.title('Matrice de confusion')
plt.show()
```
