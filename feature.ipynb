```python
import pandas as pd

def create_feature(data):
    
    # 1. Calcul de la différence de classement entre les deux joueurs
    data['Ranking_diff'] = data['Rank_1'] - data['Rank_2']
    
    # 2. Calcul de la différence de points entre les deux joueurs
    data['Point_diff'] = data['Pts_1'] - data['Pts_2']
    
    # 3. Calcul de la différence concernat la côte 
    data['Odd_diff'] = data['Odd_1'] - data['Odd_2']
    
    return data    
    
```

```python
if __name__ == '__main__':
    
    # 1. Chargement des données
    data = pd.read_csv('atp_tennis.csv')
    
    # Identification d es colonnes catégorielles (non numériques)
    categorical_columns = ['Tournament', 'Series', 'Court', 'Surface', 'Round', 'Player_1', 'Player_2', 'Date', 'Score'] 

    # Encodage one-hot des colonnes catégorielles
    data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
    
    # 2. Création des features
    data = create_feature(data)
    
    # 3. Sauvegarde des données
    data.to_csv('atp_tennis_feature.csv', index=False)
```
